```python
import pandas as pd
```

## 1. Загружаем JSON файл в dataframe

```python
df = pd.read_json('../ex02/auto.json')
df.set_index('CarNumber', inplace=True)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.float_format', '{:.2f}'.format)
pd.set_option('display.precision', 2)
def print_df(df, max_rows=None):
    with pd.option_context('display.max_rows', max_rows):
        print(df)

print("Первые 5 строк данных:")
print(df.head())
print("\nИнформация о данных:")
print(df.info())
```

## 2. Выборка

### Штрафы > 2100

```python
filtered_df_1 = df[df['Fines'] > 2100]
print(f"Всего строк: {len(filtered_df_1)}")
pd.set_option('display.max_rows', None)
print(filtered_df_1)
pd.reset_option('display.max_rows')
print("\n" + "="*80 + "\n")
```

### Штрафы > 2100 и возврат = 2

```python
filtered_df_2 = df[(df['Fines'] > 2100) & (df['Refund'] == 2.0)]
print(f"Всего строк: {len(filtered_df_2)}")
pd.set_option('display.max_rows', None)
print(filtered_df_2)
pd.reset_option('display.max_rows')
print("\n" + "="*80 + "\n")
```

### Модели из списка 'Focus' и 'Corolla'

```python
models_list = ['Focus', 'Corolla']
filtered_df_3 = df[df['Model'].isin(models_list)]
print(f"Всего строк: {len(filtered_df_3)}")
pd.set_option('display.max_rows', None)
print(filtered_df_3)
pd.reset_option('display.max_rows')
print("\n" + "="*80 + "\n")
```

### Строки с указанными номерами автомобилей

```python
car_numbers = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
filtered_df_4 = df[df.index.isin(car_numbers)]
print(f"Всего строк: {len(filtered_df_4)}")
pd.set_option('display.max_rows', None)
print(filtered_df_4)
pd.reset_option('display.max_rows')
print("\n" + "="*80 + "\n")
```

## 2. Агрегации по марке и модели

### Медианные штрафы по марке

```python
median_by_make = df.groupby('Make')['Fines'].median().round(2)
pd.set_option('display.max_rows', None)
print("Медианные штрафы по марке (округлено до 2 знаков):")
print(median_by_make)
pd.reset_option('display.max_rows')
print("\n" + "="*80 + "\n")
```

### Медианные штрафы по марке и модели

```python
median_by_make_model = df.groupby(['Make', 'Model'], dropna=False)['Fines'].median()
pd.set_option('display.max_rows', None)
print(median_by_make_model)
pd.reset_option('display.max_rows')
print("\n" + "="*80 + "\n")
```

### Количество штрафов по марке и модели

```python
count_by_make_model = df.groupby(['Make', 'Model'], dropna=False).size()
pd.set_option('display.max_rows', None)
print(count_by_make_model)
pd.reset_option('display.max_rows')
print("\n" + "="*80 + "\n")
```

### Минимальные и максимальные штрафы по марке и модели

```python
min_max_by_make_model = df.groupby(['Make', 'Model'], dropna=False)['Fines'].agg(['min', 'max'])
print(min_max_by_make_model)
print("\n" + "="*80)
```

### Стандартное отклонение штрафов по марке и модели

```python
std_by_make_model = df.groupby(['Make', 'Model'], dropna=False)['Fines'].std()
print(std_by_make_model)
print("\n" + "="*80 + "\n")
```

## 4. Агрегации по номерам автомобилей

### Номера автомобилей, сгруппированные по количеству штрафов

```python
fines_count_by_car = df.groupby(df.index).size().sort_values(ascending=False)
print(f"Всего уникальных автомобилей: {len(fines_count_by_car)}")
pd.set_option('display.max_rows', None)
print(fines_count_by_car)
pd.reset_option('display.max_rows')
top_1_car_count = fines_count_by_car.index[0]
print("\n" + "="*80 + "\n")
```

### Все строки для автомобиля с максимальным количеством штрафов

```python
top_car_df = df.loc[[top_1_car_count]]
print(f"Количество строк для этого автомобиля: {len(top_car_df)}")
pd.set_option('display.max_rows', None)
print(top_car_df)
pd.reset_option('display.max_rows')
print("\n" + "="*80 + "\n")
```

### Номера автомобилей, сгруппированные по сумме штрафов

```python
fines_sum_by_car = df.groupby(df.index)['Fines'].sum().sort_values(ascending=False)
print(f"Всего уникальных автомобилей: {len(fines_sum_by_car)}")
pd.set_option('display.max_rows', None)
print(fines_sum_by_car)
pd.reset_option('display.max_rows')
top_1_car_sum = fines_sum_by_car.index[0]
print("\n" + "="*80 + "\n")
```

### Все строки для автомобиля с максимальной суммой штрафов

```python
top_sum_car_df = df.loc[[top_1_car_sum]]
print(f"Количество строк для этого автомобиля: {len(top_sum_car_df)}")
pd.set_option('display.max_rows', None)
print(top_sum_car_df)
pd.reset_option('display.max_rows')
print("\n" + "="*80 + "\n")
```

### Проверяем, есть ли номера автомобилей, связанные с разными моделями

```python
car_model_counts = df.reset_index().groupby('CarNumber')['Model'].nunique()
cars_with_multiple_models = car_model_counts[car_model_counts > 1]
print(cars_with_multiple_models)
if len(cars_with_multiple_models) > 0:
    print(f"\nНайдено {len(cars_with_multiple_models)} автомобилей с разными моделями:")
    for car in cars_with_multiple_models.index:
        print(f"\nАвтомобиль {car}:")
        print(df.loc[[car]][['Make', 'Model']])
else:
    print("\nНет автомобилей, связанных с разными моделями.")
```
